```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# This allows one easily display the plots
%matplotlib inline
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

url = "https://www.hubertiming.com/results/2017GPTR10K"
html = urlopen(url)

soup = BeautifulSoup(html, 'lxml')

type(soup)
```

```python
#Get the title

title = soup.title

print(title)
```

```python
#Print out the text

text = soup.get_text()
# - this gives you the html of the page
print(soup.text) 
```

```python
soup.find_all('a')
```

```python
all_links = soup.find_all("a")

for link in all_links:
    print(link.get("href"))
```

```python
# Print the first ten rows only

rows = soup.find_all("tr")
print(rows[:10])
```

```python
for row in rows[:19]:
    row_td = row.find_all("td")

    
print(row_td)
type(row_td)
```

```python
#Removing the html tags for a cleaner feel

str_cells = str(row_td)

cleanText = BeautifulSoup(str_cells, "lxml").get_text()

print(cleanText)
```

```python
list_rows = []

for row in rows:
    bells = row.find_all("td")
    str_bells = str(bells)
    
    cleanTextOne = BeautifulSoup(str_bells, "lxml").get_text()
    
    list_rows.append(cleanTextOne)
```

```python
#Converting the list into a dataframe and getting a quick view of the first 10 rows using Pandas

df = pd.DataFrame(list_rows)

df.head(10)
```

```python
# The Dataframe is not in the format we want, so we clean it up a bit.We would split the 0 column into multiple columns at the comma position. This is accomplished by using the str.split() method

df1 = df[0].str.split(",", expand = True)
df1.head(10)
```

```python
# Good. There is still work to be done though. We are going to remove the square brackets on the first column

df1[0] = df1[0].str.strip("[")
df1[0] = df1[0].str.strip("]")

df1.head(10)
```

```python
#The table is missing table headers. Lets find them

col_labels = soup.find_all("th")

#Clean it up so we only get clean text 

all_header = []

col_str = str(col_labels)

cleanTextTwo = BeautifulSoup(col_str, "lxml").get_text()

all_header.append(cleanTextTwo)

print(all_header)
```

```python
df2 = pd.DataFrame(all_header)
df2.head()
```

```python
# We split this into multiple columns again

df3 = df2[0].str.split(",", expand = True)
df3.head()
```

```python
# We need to concatenate both dataframes

frames = [df3, df1]

df4 = pd.concat(frames)

df4.head(10)
```

```python
#As we can see, the header is on the first row of the table, not the head. To change that:

df5 = df4.rename(columns=df4.iloc[0])

df5.head()
```

```python
# At this point, the table is almost properly formatted. For analysis, one can get an overview of the data as shown below

df5.info()
df5.shape
```

```python
#We are going to drop all rows with any missing value

df6 = df5.dropna(axis=0, how = "any")
```

```python
# Notice how the table header is replicated as the first row in df5? We would drop that too 
df7 = df6.drop(df6.index[0])
df7.head()
```

```python
# We would perform more data cleaning by renaming the "[Place" and"Team]" columns.

df7.rename(columns= {"[Place" : "Place"}, inplace = True)
df7.rename(columns= {" Team]" : "Team"}, inplace = True)
df7.head()
```

```python
#Removing the closing bracket for cells in the "Team" column

df7["Team"] = df7["Team"].str.strip("]")
df7.head()
```

```python
# Data Analysis and Visualization
#First question, what was the average finish time (in minutes) for the runners?

time_list = df7[" Chip Time"].tolist()

time_mins = []

for i in time_list:
    timeBreakDown = i.split(":")
    
    if len(timeBreakDown) == 3:
        h = timeBreakDown[0]
        m = timeBreakDown[1]
        s = timeBreakDown[2]
    else:
        h = 0
        m = timeBreakDown[0]
        s = timeBreakDown[1]
        
    math = (int(h)*3600 + int(m)*60 + int(s))/60
    time_mins.append(math)
# print(time_mins)

#Converting the list back to a dataframe and creating a column, "Runner_mins" for runner chip times expressed in just minutes.
df7['Runner_mins'] = time_mins
df7.head()

    
    
```

```python
#This line of code shows how to calculate statistics for numeric columns only in the dataframe. The only numeric column we have is the Runner_mins one.

df7.describe(include=[np.number])
```

```python
#Interestingly, the average chip time for all runners was ~60 mins. The fastest 10K runner finished in 36.35 mins (min). And the slowest runner finished in 101.30 minutes.

#A boxplot is another useful tool to visualize summary statistics(maximum, minimum, medium, first quartile, second quartile, third quartile,including outliers.)

from pylab import rcParams
rcParams['figure.figsize'] = 15,5

df7.boxplot(column = 'Runner_mins')
plt.grid(True, axis='y')
plt.ylabel('Chip Time')

plt.xticks([1], ['Runners'])
```

```python
#The second question is: Did the runners' finish times follow a normal distribution?

x = df7['Runner_mins']

ax = sns.distplot(x, hist = True, kde = True, rug = False, color ='m', bins = 25, hist_kws = {'edgecolor':'black'})

plt.show()
```

```python
#The third question deals with whether there were any perfomance differences between males and females of various age groups.

f_fuko = df7.loc[df7[' Gender']== ' F']['Runner_mins']
m_fuko = df7.loc[df7[' Gender']==' M']['Runner_mins']

sns.distplot(f_fuko, hist=True, kde=True, rug=False, hist_kws = {'edgecolor':'black'}, label='Female')
sns.distplot(m_fuko, hist=True, kde=True, rug=False, hist_kws = {'edgecolor':'black'}, label='Male')

plt.legend()
```

```python
#The distribution indicates that females were slower than males on average. We would use the groupby method to compute summary statistics for males and females separately as shown below

g_stats = df7.groupby(" Gender", as_index=True).describe()

print(g_stats)
```

```python
#The average chip time for all females and males was ~66 mins and ~58 mins respectively. Below is a side-by-side boxplot comparisonn of male and female finish times.

df7.boxplot(column='Runner_mins', by= ' Gender')
plt.ylabel('Chip Time')
plt.suptitle("")
```
